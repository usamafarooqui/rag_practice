```python
from langchain_community.document_loaders import TextLoader
### textloader text ko document structure krnay mein help krta hai
loader = TextLoader("../data/text_files/python_intro.txt" , encoding="utf-8")
document = loader.load()
print(document)
```

```python
### Directory Loader 
from langchain_community.document_loaders import DirectoryLoader

dir_loader=DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt", ##patter tp match files
    loader_cls=TextLoader, ##loader class to use
    loader_kwargs={'encoding':'utf-8'},
    show_progress=False
)

documents = dir_loader.load()
print(documents)
```

```python
### Directory Loader for pdf (dono functions ko 1 sath bhi kr saktay thay just loader_ccls mein list de detay dono loaders ki)
from langchain_community.document_loaders import PyMuPDFLoader , PyPDFLoader

dir_loader=DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf", ##patter tp match files
    loader_cls=PyMuPDFLoader, ##loader class to use
    # loader_kwargs={'encoding':'utf-8'},
    show_progress=False
)

pdfdocuments = dir_loader.load()
print(pdfdocuments)
```
